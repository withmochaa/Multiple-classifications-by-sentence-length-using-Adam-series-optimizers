```python
count_기관형=0
count_s=0
count_sentence=0

for i in range(data_row_기관사칭형.shape[0]):
    A=len(data_row_기관사칭형.iloc[i,0])
    for j in range(A):
        if data_row_기관사칭형.iloc[i,0][j]==',':
            count_s=count_s+1
        if count_s==3:
            count_sentence=count_sentence+1
            count_s=0
                  
print(count_sentence)
```

```python
기관사칭형=pd.DataFrame(index=range(count_sentence),columns=['sentence'])
```

```python
count=0
B=[]
for i in range(data_row_기관사칭형.shape[0]):
    A=len(data_row_기관사칭형.iloc[i,0])
    for j in range(A):
        if data_row_기관사칭형.iloc[i,0][j]!=','and data_row_기관사칭형.iloc[i,0][j]!='\'' and data_row_기관사칭형.iloc[i,0][j]!='[' and data_row_기관사칭형.iloc[i,0][j]!=']':
            B.append(data_row_기관사칭형.iloc[i,0][j])
        if data_row_기관사칭형.iloc[i,0][j]==',':
            count_s=count_s+1
        if count_s==3:
            count_s=0
            str = ''.join(B)
            기관사칭형.iloc[count,0]=str
            count=count+1
            B=[]
```
